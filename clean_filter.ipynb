# **UK Product Filter**

Since this project will be focusing on the UK public, it makes sense to focus on using products available in the UK.

The output is a dataframe with around 115,000 products.

```python
# filtering products without countries information
df_off_non_nan_countries = df_off[df_off['countries'].notna()]

# transforming the countries information to lowercase
df_off_non_nan_countries['countries'] = df_off_non_nan_countries['countries'].str.lower()
```

```python
# using different variations of referring to the united kingdom (and scotland) to filter the products

uk_variations = ['united kingdom', 'united-kingdom', 'uk', 'gb', 'england', 'scotland']
uk_pattern = '|'.join(uk_variations)
df_uk = df_off_non_nan_countries[df_off_non_nan_countries['countries'].str.contains(uk_pattern)]
df_uk
```

```python
# exporting the filtered data to a csv file
df_uk.to_csv('df_uk.csv', index=False)
```
