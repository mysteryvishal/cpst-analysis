```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('Google Trends UPF.csv', skiprows=2)

# Clean the data
df['year'] = pd.to_datetime(df['Week']).dt.year
df.drop('Week', axis=1, inplace=True)

# Group by year and sum
df = df.groupby('year').sum().sum(axis=1)
```

```python
df.plot(
    title='Google Trends search volume for related phrases to ultra-processed foods',
    xlabel='Year',
    ylabel='Search volume',
)
```
